```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
#balancing the dataset

df_yes = df[df['cancer']==1]
df_no = df[df['cancer']==0]
```

```python
#sample 1158 from df_no

df_no_sampled = df_no.sample(1158,replace=False,ignore_index=True)
```

```python
df_no_sampled
```

```python
#merging the df_yes and df_no_sampled into one dataframe

df_balanced = pd.concat([df_yes,df_no_sampled],axis=0).reset_index(drop=True)
```

```python
#adding a column for image name
def get_image_name(row):

    return str(row['patient_id']) + "_" + str(row['image_id']) + ".png"
```

```python
df_balanced['image_name'] = df_balanced.apply(get_image_name,axis=1)
```

```python
#converting the datatype of cancer column from int to str because it is expected by keras model which we will be using later

df_balanced['cancer'] = df_balanced['cancer'].astype(str)
```

```python
#shuffling the rows in the balanced df because otherwise it feels illegal

df_balanced = df_balanced.sample(frac=1).reset_index(drop=True)
```

```python
df_balanced
```

```python
#our dataframe is ready, now we can store it as csv and later use it for further tasks.
df_balanced.to_csv("balanced_data.csv")
```
